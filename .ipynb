```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def getdata():

    df = pd.read_csv('Causes of death.csv')#to fetch the data from csv file

    afg_df = df[df['Code'].eq('AFG')]#Take data whcih only contain country Name Afganistan with country code AFG
    afg_df.dropna#for dropping Null Values from the data
    afg_df.dtypes#Checking datatypes of all columns from the data

    return afg_df

def lineplot():
    '''
    This function showing line plot for death causes in afganistan due to diseases name as Nutritional Deficiencies,Diabetes Mellitus,Chronic Respiratory Diseases,Road Injuries 
    '''
    plt.figure(dpi=100)
    plt.title("Line Plot For Causes of Death In Afganistan")#this function use for giving title name of graph

    plt.plot(afg_df['Year'], afg_df['Nutritional Deficiencies'],'tab:green',label='Nutritional Deficiencies')
    plt.plot(afg_df['Year'], afg_df['Diabetes Mellitus'],'tab:red',label='Diabetes Mellitus')
    plt.plot(afg_df['Year'], afg_df['Chronic Respiratory Diseases'],label='Chronic Respiratory Diseases')
    plt.plot(afg_df['Year'], afg_df['Road Injuries'],'tab:orange',label='Road Injuries')

    plt.xlabel('Years')#lable on x axis
    plt.ylabel('Number Of Deaths')#lable on y axis
    plt.legend(loc='upper right')#this function use for giving label in graph
    plt.yticks(np.arange(1000, 15000, 2000))
    plt.xlim(1990,2019)
    plt.show()

def piechart():
    '''
    This function showing Piechart for death causes in afganistan due to diseases name as Malaria,Nutritional Deficiencies,Drowning,Interpersonal Violence,Meningitis 
    '''
    parkinsons_disease = np.sum(afg_df['Malaria'])
    nutritional_deficiencies = np.sum(afg_df['Nutritional Deficiencies'])
    drowning = np.sum(afg_df['Drowning'])
    interpersonal_violence = np.sum(afg_df['Interpersonal Violence'])
    meningitis = np.sum(afg_df['Meningitis'])
    meningitis
    total = parkinsons_disease + nutritional_deficiencies + drowning + interpersonal_violence + meningitis

    par_dis = parkinsons_disease / total*100.0
    nutri_defi = nutritional_deficiencies / total*100.0
    drow = drowning / total*100.0
    inter_vio = interpersonal_violence / total*100.0
    meni = meningitis / total*100.0

    deaths = np.array([nutri_defi, drow, inter_vio, meni])
    diseases = [ 'Nutritional Deficiencies', 'Drowning', 'Interpersonal Violence', 'Meningitis']
    plt.figure(dpi=110)
    plt.title("Pie Chart For Causes of Death In Afganistan")
    plt.xlim(1990,2019)
    plt.pie(deaths,labels = diseases, autopct='%1.1f%%', shadow=True, explode= (0,0,0,0.1))
    plt.show()
    
def barplotgroup():
    '''
    This function showing Bar plot for death causes in afganistan due to diseases name as Digestive Diseases,Fire, Acute Hepatitis,Poisonings 
    '''
    afg_df = df.loc[70:75]#size for the plot
    num = np.arange(6)
    width = 0.3
    years = afg_df['Year'].tolist()

    plt.figure(dpi=120)
    plt.title('Bar chart For Causes of Death In Afganistan')#title for the plot
    plt.bar(num, afg_df['Digestive Diseases'], width, label='Digestive Diseases')
    plt.bar(num+0.2, afg_df['Fire, Heat, and Hot Substances'], width, label='Fire, Heat, and Hot Substances')
    plt.bar(num-0.2, afg_df['Acute Hepatitis'], width, label='Acute Hepatitis')
    plt.bar(num+0.4, afg_df['Poisonings'], width, label='Poisonings')
    plt.xticks(num, years)
    plt.yticks(np.arange(100,6000,1000))
    plt.xlabel('Years')#showing years on x axis
    plt.ylabel('Number Of Deaths')#showing years on y axis
    plt.legend(loc='upper right',bbox_to_anchor=(1.4, 0, 0.13, 1.4))
    plt.show()



afg_df = getdata()
lineplot()
piechart()
barplotgroup()
afg_df
#print(afg_df)


```
